```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
# Step 3: Display basic information about the dataset
print("Dataset shape:", titanic.shape)
print("\nFirst 5 rows of the dataset:")
display(titanic.head())
```

```python
# Step 4: Check for missing values
print("\nMissing values in each column:")
print(titanic.isnull().sum())
```

```python
# Step 5: Basic statistical summary
print("\nStatistical summary of numerical features:")
display(titanic.describe())
```

```python
# Step 6: Plot the distribution of ticket fare using histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x='fare', kde=True, bins=30)
plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare (in $)')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Step 7: Further exploratory analysis - fare by passenger class
plt.figure(figsize=(12, 6))
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd)')
plt.ylabel('Fare (in $)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Step 8: Explore relationship between survival and fare
plt.figure(figsize=(12, 6))
sns.violinplot(x='survived', y='fare', data=titanic)
plt.title('Fare Distribution by Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare (in $)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Step 9: Visualization of fare by gender and class
plt.figure(figsize=(14, 7))
sns.boxplot(x='pclass', y='fare', hue='sex', data=titanic)
plt.title('Fare Distribution by Class and Gender')
plt.xlabel('Passenger Class')
plt.ylabel('Fare (in $)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Step 10: Create a more detailed histogram of fare with multiple groups
plt.figure(figsize=(12, 8))
sns.histplot(data=titanic, x='fare', hue='survived', multiple='stack', bins=30)
plt.title('Distribution of Ticket Fare by Survival Status')
plt.xlabel('Fare (in $)')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset (if you're using seaborn's built-in dataset)
titanic = sns.load_dataset('titanic')

# Plot
sns.countplot(x='survived', data=titanic)
plt.title("Survival Count")
plt.show()
```

```python
# Survival by gender
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival by Gender")
plt.show()
```

```python
# Survival by class
sns.countplot(x='class', hue='survived', data=titanic)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
# Age distribution
sns.histplot(titanic['age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# Plot histogram of fare
sns.histplot(titanic['fare'], bins=30, kde=True)
plt.title("Ticket Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.show()
```
