```python
# Calculate the mean of the 'crim' column (crime rate) from the dataframe
mean_value = df['crim'].mean()

# Print the mean value of the 'crim' column
print('Mean value of CRIM:', mean_value)
```

```python
target_feature = 'medv'
```

```python
x = df.drop(target_feature, axis=1)  # All columns except 'medv'
y = df[target_feature]  # Only the 'medv' column as the target
```

```python
x.head()
```

```python
y.head()
```

```python
# Import the train_test_split function from sklearn.model_selection
from sklearn.model_selection import train_test_split

# Split the dataset into training and testing sets
# x contains the features, y contains the target variable
# 80% of data is used for training, 20% for testing, with a fixed random state for reproducibility
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# Import the LinearRegression model from sklearn.linear_model
from sklearn.linear_model import LinearRegression

# Create an instance of the LinearRegression model
regression = LinearRegression()
```

```python
# Fit the regression model using the training data (x_train and y_train)
regression.fit(x_train, y_train)
```

```python
# Calculate the R-squared score (coefficient of determination) on the training set
train_score = round(regression.score(x_train, y_train) * 100, 2)

# Print the train score as a percentage
print('Train score of Linear Regression:', train_score)
```

```python
# Print the coefficients (weights) assigned to each feature by the linear regression model
print('Coefficients: ', regression.coef_)
```

```python
# Predict the target values (y) using the test data (x_test) based on the trained model
predictions = regression.predict(x_test)
```

```python
predictions
```

```python
# Create a scatter plot to compare the actual values (y_test) with the predicted values (predictions)
plt.scatter(y_test, predictions)

# Label the x-axis as the actual values of the target variable
plt.xlabel('Y Test')

# Label the y-axis as the predicted values of the target variable
plt.ylabel('Predicted Y')
```

```python
# Import the r2_score function from sklearn.metrics to evaluate model performance
from sklearn.metrics import r2_score

# Calculate the r2 score by comparing the actual test values (y_test) and predicted values (predictions)
score = round(r2_score(y_test, predictions) * 100, 2)

# Print the r2 score, which indicates the proportion of variance explained by the model
print("r_2 score:", score)
```

```python
# Calculate the R² score for the test data and round it to two decimal places
# The .score method computes the coefficient of determination (R² score) for the model
test_score = round(regression.score(x_test, y_test) * 100, 2)

# Print the test R² score, indicating how well the model performs on the test data
print('Test score of Linear Regression:', test_score)
```

```python
from sklearn import metrics

# Calculate and print the Mean Absolute Error (MAE) on the test data
# MAE measures the average magnitude of the errors in a set of predictions, without considering their direction
print('Mean Absolute Error on test data of Linear Regression: ', metrics.mean_absolute_error(y_test, predictions))

# Calculate and print the Mean Squared Error (MSE) on the test data
# MSE measures the average squared difference between the estimated values and the actual value
print('Mean Squared Error on test data of Linear Regression: ', metrics.mean_squared_error(y_test, predictions))

# Calculate and print the Root Mean Squared Error (RMSE) on the test data
# RMSE is the square root of MSE and gives the error in the same units as the target variable
print('Root Mean Squared Error on test data of Linear Regression: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
