```python
# Optional: Check and convert columns to numeric if needed
df["arrivals_in_qtl"] = pd.to_numeric(df["arrivals_in_qtl"], errors="coerce")
df["min_price"] = pd.to_numeric(df["min_price"], errors="coerce")
df["max_price"] = pd.to_numeric(df["max_price"], errors="coerce")
df["modal_price"] = pd.to_numeric(df["modal_price"], errors="coerce")

# Step 1: Calculate averages (ignores NaNs automatically)
avg_arrivals = df["arrivals_in_qtl"].mean()
avg_min = df["min_price"].mean()
avg_max = df["max_price"].mean()
avg_modal = df["modal_price"].mean()

# Step 2: Print or store the results
print("Average Arrivals (in qtl):", avg_arrivals)
print("Average Min Price:", avg_min)
print("Average Max Price:", avg_max)
print("Average Modal Price:", avg_modal)
```

```python
# 1. Convert the column to float (handles non-numeric safely)
df["arrivals_in_qtl"] = pd.to_numeric(df["arrivals_in_qtl"], errors='coerce')

# 2. Calculate the average (mean), ignoring NaNs
avg_arrivals = df["arrivals_in_qtl"].mean()

# 3. Replace NaN values with the average
df["arrivals_in_qtl"] = df["arrivals_in_qtl"].replace(np.nan, avg_arrivals)

# Optional: Print the average to confirm
print("Average Arrivals (in qtl):", avg_arrivals)
```

```python
# Step 1: Convert the 'arrivals_in_qtl' column to float (just in case it's not already numeric)
df["min_price"] = df["min_price"].astype("float")

# Step 2: Calculate the mean of the column, ignoring NaN values
avg_arrivals = df["min_price"].mean()

# Step 3: Replace NaN values with the mean
# Note: Don't use inplace=True here to avoid chained assignment warning
df["min_price"] = df["min_price"].replace(np.nan, avg_arrivals)

# Optional: Print the average for verification
print("Average min_price:", avg_arrivals)
```

```python
# Replace the missing values in the "arrivals_in_qtl" column by the most frequent value
df["arrivals_in_qtl"] = df["arrivals_in_qtl"].replace(np.nan, df["arrivals_in_qtl"].mode()[0])

# Simply drop the whole row with NaN in "district_name" column
df = df.dropna(subset=["district_name"], axis=0)

# Reset the index, because we dropped some rows
df = df.reset_index(drop=True)

# Check for missing values in the dataset
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numeric columns (e.g., "arrivals_in_qtl", "min_price", "max_price")
df[["arrivals_in_qtl", "min_price", "max_price"]] = scaler.fit_transform(df[["arrivals_in_qtl", "min_price", "max_price"]])
```
