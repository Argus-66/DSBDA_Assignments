```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
sns.set_style("whitegrid")

# Assuming your iris dataset is loaded as 'iris'
sns.FacetGrid(iris, hue="variety", height=7).map(plt.scatter, "sepal.length", "sepal.width").add_legend()

# Display the plot
plt.show()
```

```python
# Filter rows where the species is 'Iris-setosa'
setosa_data = iris[iris['variety'] == 'Setosa']

# Filter rows where the species is 'Iris-versicolor'
versicolor_data = iris[iris['variety'] == 'Virginica']

# Display the first few rows of each dataset
print("Setosa Data:")
print(setosa_data.head())

print("Versicolor Data:")
print(versicolor_data.head())
```

```python
setosa_data
```

```python
versicolor_data
```

```python
setosa_stats = {
    'percentile': np.percentile(setosa_data['sepal.length'], 50),  # Corrected column name
    'mean': np.mean(setosa_data['sepal.length']),
    'std_dev': np.std(setosa_data['sepal.length'])
}

print(setosa_stats)
```

```python
versicolor_stats = {
    'percentile': np.percentile(versicolor_data['sepal.length'], 50),  # Corrected column name
    'mean': np.mean(versicolor_data['sepal.length']),
    'std_dev': np.std(versicolor_data['sepal.length'])
}

print(versicolor_stats)
```

```python
print('Statistics for Iris Setosa:')
print(setosa_stats)
print('')

print('Statistics for Iris Versicolor:')
print(versicolor_stats)
```
